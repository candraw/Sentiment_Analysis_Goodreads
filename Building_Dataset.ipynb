```python
from google_play_scraper import app
import pandas as pd
import numpy as np
```

```python

from google_play_scraper import Sort, reviews
result, continuation_token = reviews(
    'com.goodreads',
    lang='id', # defaults to 'en'
    country='id', # defaults to 'us'
    sort=Sort.NEWEST, # defaults to Sort.MOST_RELEVANT you can use Sort.NEWEST to get newst reviews
    count=999, # defaults to 100
    filter_score_with= 1  # defaults to None(means all score) Use 1 or 2 or 3 or 4 or 5 to select certain score
)
```

```python
review1 = review1 = pd.DataFrame(np.array(result),columns=['review'])
review1 = review1.join(pd.DataFrame(review1.pop('review').tolist()))
review1.tail()
```
